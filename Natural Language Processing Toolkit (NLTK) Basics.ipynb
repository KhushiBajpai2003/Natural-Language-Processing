```python
def text_lowercase(text):
    return text.lower()
```

```python
from nltk.corpus import brown

# Load the Brown Corpus
corpus = brown

# Access the words from the first chapter
chapter_words = corpus.words(categories='news')[:100]  # Adjust the number as needed

# Convert the list of words to a string
chapter_text = ' '.join(chapter_words)

# Print or use the chapter_text variable as needed
print(chapter_text)
```

```python
text_lowercase(chapter_text)
```

```python
import string

def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)

# Assuming chapter_text is the text you obtained from the Brown Corpus
cleaned_text = remove_punctuation(chapter_text)

# Now you can use cleaned_text as needed
print(cleaned_text)
```

```python


def remove_whitespaces(text):
    return " ".join(text.split())

# Assuming chapter_text is the text you obtained from the Brown Corpus
cleaned_text = remove_whitespaces(chapter_text)

# Now you can use cleaned_text as needed
print(cleaned_text)
```
